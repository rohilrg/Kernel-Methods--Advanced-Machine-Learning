```python
from SVDD_definitions import one_class_dataset_generator,svdd
from sklearn import svm
```

```python
X_array,X=one_class_dataset_generator(n_samples=300,number_of_outliers=15,plot=True)
```

# Linear Kernel Results

```python
for i in range(1,6,1):
    i= i/10
    svdd(X_array,X,Constant_for_SVDD=i,type_of_kernel='linear')
```

# RBF-Kernel Results

```python
for i in range(1,6,1):
    i= i/10
    svdd(X_array,X,Constant_for_SVDD=i,type_of_kernel='rbf',gamma_rbf=0.001)
```

# Polynomial Kernel Results

```python
for i in range(1,6,1):
    i= i/10
    svdd(X_array,X,Constant_for_SVDD=i,type_of_kernel='poly',gamma_poly=0.001,coeff_poly=1, degree_poly=3)
```
