```python
from Kernel_PCA_definitions import swiss_roll_dataset,classic_pca,kernel_pca,plot_graph
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
X,color = swiss_roll_dataset(number_of_samples=1200,plot=True)
```

## Classic PCA implementation

```python
X_transformed_2=classic_pca(X,color,n_components=2)
```

```python
X_transformed_1=classic_pca(X,color,n_components=1)
```

## Kernel PCA Implementation

### Kernel PCA with RBF Kernel with first 2 principle components

```python
for gamma in range(1,10,2):
    start_time = time.time()
    gamma= gamma/100
    X_pc = kernel_pca(X, n_components=2,type_of_kernel='rbf',gamma=gamma)
    plt.figure(figsize=(8, 6))
    plt.scatter(X_pc[:, 0], X_pc[:, 1], c=color, cmap=plt.cm.rainbow)
    plt.title('First 2 principal components after RBF Kernel PCA')
    plt.text(-0.14, 0.14, 'gamma = {}'.format(gamma), fontsize=12)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.show()
    print("The running time for RBF Kernel with",gamma,'gamma is %s'%(time.time()-start_time),'seconds')
```

## K-PCA with RBF Kernel with first principal component

```python
for gamma in range(1,10,2):
    start_time = time.time()
    gamma= gamma/100
    X_pc = kernel_pca(X, n_components=1,type_of_kernel='rbf',gamma=gamma)

    plt.figure(figsize=(8,6))
    plt.scatter(X_pc[:,0], np.zeros((X_pc.shape[0],1)), c=color.reshape((X_pc.shape[0],1)), cmap=plt.cm.rainbow)

    plt.text(-0.125, 0.007, 'gamma = {}'.format(gamma), fontsize=12)
    plt.title('First principal component after RBF Kernel PCA')
    plt.xlabel('PC1')
    plt.show()
    print("The running time for RBF Kernel with",gamma,'gamma is %s'%(time.time()-start_time),'seconds')
```

### K-PCA with Polynomial Kernel

```python
for poly_const in np.array([0.01,0.1,1,10,100,1000,10000,100000]):
    start_time = time.time()
    X_pc = kernel_pca(X, n_components=2,type_of_kernel='poly',poly_constant=poly_const)
    plt.figure(figsize=(8, 6))
    plt.scatter(X_pc[:, 0], X_pc[:, 1], c=color, cmap=plt.cm.rainbow)
    plt.title('First 2 principal components after Polynomial Kernel PCA')
    plt.text(-0.14, 0.14, 'poly_constant = {}'.format(poly_const), fontsize=12)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.show()
    print("The running time for RBF Kernel with",poly_const,'polynomial kernel constant is %s'%(time.time()-start_time),'seconds')
```

#### K-PCA with Sigmoid Kernel

```python
for sigmoid_const in np.array([1,10,20]):
    for sigmoid_alpha in np.array([0.001,0.01]):   
        start_time = time.time()
        X_pc = kernel_pca(X, n_components=2,type_of_kernel='sigmoid',sigmoid_alpha=sigmoid_alpha,sigmoid_constant=sigmoid_const)
        plt.figure(figsize=(8, 6))
        plt.scatter(X_pc[:, 0], X_pc[:, 1], c=color, cmap=plt.cm.rainbow)
        plt.title('First 2 principal components after RBF Kernel PCA')
        plt.text(-0.14, 0.14, ('sigmoid_constant = {}'.format(sigmoid_const),'sigmoid_alpha={}'.format(sigmoid_alpha)), fontsize=12)
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.tight_layout()
        plt.show()
        
        print("The running time for RBF Kernel with",sigmoid_const,'sigmoid kernel constant and',sigmoid_alpha,'sigmoid kernel alpha is %s'%(time.time()-start_time),'seconds')
```
